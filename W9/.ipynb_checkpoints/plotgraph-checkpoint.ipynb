```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('database.csv')
```

## First graph

```python
# number of wins
winning_times = data[data['Result'] == 'win']
winning_times = winning_times.groupby(by=["Player"]).count()
print(winning_times['Result'].plot(kind='bar'))
```

## Second graph

```python
# number of draw
winning_times = data[data['Result'] == 'draw']
winning_times = winning_times.groupby(by=["Player"]).count()
print(winning_times['Result'].plot(kind='bar'))
```

## Third graph

```python
# number of occurrences
game_num = data['Game'].max()
show_num = data.groupby(by=["Player"]).count()
show_num['Percent'] = show_num['Result'] / data['Game'].max()
print(show_num['Percent'].plot(kind='pie'))
```
